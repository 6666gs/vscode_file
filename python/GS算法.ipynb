```python
import math
import numpy as np
import cv2 
import matplotlib.pyplot as plt # plt 

def Ger_Sax_algo(img, max_iter):
    h, w = img.shape
    t_am = np.sqrt(img)
    t_pm = np.random.rand(h, w)
    s_am = np.ones((h, w))
    s_pm = np.ones((h, w))
    
    t=t_am*np.exp(1j*t_pm)
    s=np.fft.ifft2(t)
 
    for iter in range(max_iter):
        s=s_am*np.exp(1j*np.angle(s))
        t=np.fft.fft2(s)
        t=t_am*np.exp(1j*np.angle(t))
        s=np.fft.ifft2(t)
        
 
    pm =np.angle(s)
    return pm

IMG=cv2.imread('A.png',0)
pm=Ger_Sax_algo(IMG,50)
h,w =IMG.shape
s_am=np.ones((h,w))
f=np.fft.fft2(s_am*np.exp(pm*1j))
I_f=abs(f)**2
plt.imshow(I_f)
```
